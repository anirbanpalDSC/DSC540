```python
# Multiple Queries

# Create function to store multiple tweets
# Insert the result in a sqlite table (data wrangling)
def store_tweet(item):
    db = dataset.connect('sqlite:///data_wrangling.db')
    table = db['tweets']
    item_json = item._json.copy()
    for k, v in item_json.items():
        if isinstance(v, dict):
            item_json[k] = str(v)
    table.insert(item_json)

# Craete tweepy object for authentication
auth = tweepy.OAuthHandler(API_KEY, API_SECRET)
auth.set_access_token(TOKEN_KEY, TOKEN_SECRET)

api = tweepy.API(auth)

# Do repeatitive query using a cursor
query = '#childlabor'
cursor = tweepy.Cursor(api.search, q=query, lang="en")

# Store data from the cursor to the sqlite table
for page in cursor.pages():
    for item in page:
        store_tweet(item)
```
