```python
from csv import reader
from datetime import datetime

# Read data file and header file
data_rdr = reader(open("../Data/unicef/mn.csv", "rt"))
header_rdr = reader(open("../Data/unicef/mn_headers.csv", "rt", encoding="utf8"))

# Create lists for data and header
data_rows = [d for d in data_rdr]
header_rows = [h for h in header_rdr]

# Check whether all fields are loaded or not
print(len(data_rows[0]))
print(len(header_rows))
```

From the lengths above, we can see that the header has additional fields (or vice versa). Therefore, those fields need to be corrected in both header and data before we merge them.

```python
# List the rows identified as having mismatched headers to the bad_rows list and then 
# identify rows to remove

bad_rows = []

for h in header_rows:
    if h[0] not in data_rows[0]:
        bad_rows.append(h)
        
for h in bad_rows:
    header_rows.remove(h)
    
print(len(header_rows))
```

```python
# Flag headers not matching the clean headers list

all_short_headers = [h[0] for h in header_rows]

for header in data_rows[0]:
    if header not in all_short_headers:
        print('mismatch!', header)
```

```python
# Stores the data rows that we are not interested in

skip_index = []
for header in data_rows[0]:
    if header not in all_short_headers:
        index = data_rows[0].index(header)
        skip_index.append(index)
```

```python
# Create a list for the intended rows, zip each row and add to array

new_data = []

for row in data_rows[1:]:
    new_row = []
    for i, d in enumerate(row):
        if i not in skip_index:
            new_row.append(d)
    new_data.append(new_row)

zipped_data = []

for drow in new_data:
    zipped_data.append(zip(header_rows, drow))
```

```python
# Format zipped_data to show in a human readable format

for x in zipped_data[0]:
    print('Question: {}\nAnswer: {}'.format(x[0], x[1]))
```

```python
# Since I am using Python 3 which does not return list from zip (like Python 2 does), 
#I am converting 'zipped_data' into a tuple called 'tpl_data' and doing the rest of the excercise

tpl_data = tuple(zip(*zipped_data))

# Formatting the tuple elements for Month, Date, Year, Hour and Minute 
# To display interview start time

start_string = '{}/{}/{} {}:{}'.format(
    tpl_data[8][20][1],tpl_data[7][20][1],tpl_data[9][20][1],
    tpl_data[13][20][1],tpl_data[14][20][1])

print(start_string)

start_time = datetime.strptime(start_string, '%m/%d/%Y %H:%M')

print(start_time)
```
