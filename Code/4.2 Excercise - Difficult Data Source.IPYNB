For this task I am using the PyPDF2 library. I cannot use `slate` since I am using `Python 3` not `Python 2`.

```python
import PyPDF2

pdfFileObj = open('Data//EN-FINAL Table 9.pdf', 'rb', 1) 
  
# creating a pdf reader object 
pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 

pdfWriter = PyPDF2.PdfFileWriter()
  
# creating a page object
pg1 = pdfReader.getPage(0) 
pg2 = pdfReader.getPage(1) 

pdfWriter.addPage(pg1)
pdfWriter.addPage(pg2)

NewPDFfilename = 'EN-FINAL Table 1 and 2.pdf'
with open(NewPDFfilename, "wb") as outputStream: #create new PDF
    pdfWriter.write(outputStream)

# extracting text from page 
print(pg1.extractText()) 

# closing the pdf file object 
pdfFileObj.close() 
```
